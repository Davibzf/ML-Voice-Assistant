Desenvolver um assistente de voz local, feito em Python, capaz de:

### üéôÔ∏èFluxo Geral de Funcionamento
1. üé§O usu√°rio fala 
2. üìùO sistema converte voz em texto (STT) 
3. üß†O texto √© interpretado pela ML (GEMINI)    
4. üó£Ô∏èO assistente responde por voz (TTS)



```python
# 1. Grava√ß√£o do √°udio do usu√°rio
# Captura de √°udio via microfone e salvamento em arquivo .wav


# Importa√ß√µes das bibliotecas de grava√ß√£o de √°udio
import pyaudio
import wave
import threading

# Configura√ß√µes de √°udio 
FORMAT = pyaudio.paInt16
CHANNELS = 1
RATE = 16000
CHUNK = 1024 
OUTPUT_FILE = "gravacao.wav"

# Inicializa√ß√£o do PyAudio
audio = pyaudio.PyAudio()

# Controle manual pelo teclado (Jupyter-friendly)
input("üé§ Pressione ENTER para come√ßar a gravar...")
print("Gravando... Pressione ENTER novamente para parar.")

# Abrindo o microfone 
stream = audio.open(
    format=FORMAT,
    channels=CHANNELS,
    rate=RATE,
    input=True,
    frames_per_buffer=CHUNK
)

# Buffer de √°udio
frames = []

# Vari√°vel de controle
stop = False

# Fun√ß√£o que espera o ENTER
def wait_stop():
    global stop
    input()
    stop = True

# Thread paralela
threading.Thread(target=wait_stop).start()

# Loop de grava√ß√£o
while not stop:
    data = stream.read(CHUNK)
    frames.append(data)

# Encerramento da grava√ß√£o
print("‚èπÔ∏è Grava√ß√£o finalizada")

stream.stop_stream()
stream.close()
audio.terminate()

# Salvando o arquivo WAV
with wave.open(OUTPUT_FILE, 'wb') as arquivo:
    arquivo.setnchannels(CHANNELS)
    arquivo.setsampwidth(audio.get_sample_size(FORMAT))
    arquivo.setframerate(RATE)
    arquivo.writeframes(b"".join(frames))

print(f"‚úÖ √Åudio salvo!")
```

```python
# 2. Transcri√ß√£o de √°udio para texto
# Uso do modelo Whisper (tiny) para converter o √°udio gravado em texto

# Impota√ßao da STT
import whisper 

# Modelo Tiny versao mais leve e rapida do ML de reconhecimento e transcri√ß√£o de fala, criado pela OpenAI
modelo = whisper.load_model('tiny')

# Par√¢metros de importa√ß√£o do √°udio
resultado = modelo.transcribe('{}'.format('gravacao.wav'), language ='pt')

# Par√¢metros de salvamento do texto transcrito
with open("transcricao.txt", "w", encoding="utf-8") as f:
    f.write(resultado["text"])
```

```python
# 3. Interpreta√ß√£o com ia (gemini)
# Envio do texto transcrito para a API do Google Gemini e obten√ß√£o da resposta


# Importa√ßao do gemini 
import google.generativeai as genai

# Importa√ßao do texto e atribui√ß√£o a variavel {per}
with open('transcricao.txt', 'r') as arquivo:
    conteudo = arquivo.read()
    per = f"""
Voc√™ √© um assistente objetivo.
Responda de forma curta.
Pergunta: {conteudo}
"""


# Chave api
import os
genai.configure(api_key=os.getenv("GEMINI_API_KEY"))

# Modelo seleciona do gemini
model = genai.GenerativeModel("models/gemini-flash-lite-latest")

# Pergunta
response = model.generate_content(
    per
)

resp = response.text
print(resp)
```

```python
# 4. Convers√£o de texto para √°udio (tts)
# Leitura da resposta gerada e convers√£o em voz com pyttsx3


# Importar a biblioteca(TTS)
import pyttsx3
# Removedor de caracteres especiais 
import re

# Inicializa√ß√£o da engine
engine = pyttsx3.init()

# Configura√ß√£o 
engine.setProperty('rate', 125)
engine.setProperty('volume', 1.0)



engine.say(resp)



engine.runAndWait()



```
